```python
#importing all the necessary libraries
import cv2
import matplotlib.pyplot as plt
import dlib
from imutils import face_utils
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
#setting path and declare Cascade classifiers
cascPath = "venv/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml"
eyePath = "venv/Lib/site-packages/cv2/data/haarcascade_eye.xml"
smilePath = "venv/Lib/site-packages/cv2/data/haarcascade_smile.xml"
```

```python
#assigning the clasifiers to different variables
faceCascade = cv2.CascadeClassifier(cascPath)
eyeCascade = cv2.CascadeClassifier(eyePath)
smileCascade = cv2.CascadeClassifier(smilePath)
```

```python
#loading a single image for trial
gray = cv2.imread('image/anna.jpg', 0)
plt.figure(figsize=(10,6))
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
# Detect faces
faces = faceCascade.detectMultiScale(
gray,
scaleFactor=1.1,
minNeighbors=5,
flags=cv2.CASCADE_SCALE_IMAGE
)
```

```python
# For each face
for (x, y, w, h) in faces: 
    # Draw rectangle around the face
    cv2.rectangle(gray, (x, y), (x+w, y+h), (255, 255, 255), 3)
```
